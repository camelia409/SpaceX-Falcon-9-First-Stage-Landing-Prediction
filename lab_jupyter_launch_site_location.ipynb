Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# Define the closest coastline coordinates (replace with actual values from MousePosition)
coastline_lat, coastline_lon = 28.56367, -80.57163  

# Define the launch site coordinates (replace with actual values from spacex_df)
launch_site_lat, launch_site_lon = 28.563197, -80.576820  

# Compute distance
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
print(f"Distance from launch site to coastline: {distance_coastline:.2f} km")
```

```python
import folium
from folium.plugins import MarkerCluster
from folium.features import DivIcon

# Create a folium marker for the coastline
coastline_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=folium.Icon(color="blue", icon="cloud"),
    popup="Closest Coastline"
)

# Add marker to map
site_map.add_child(coastline_marker)
site_map
# Create a text label for the distance
distance_marker = folium.Marker(
    [ (launch_site_lat + coastline_lat) / 2, (launch_site_lon + coastline_lon) / 2 ],  # Midpoint for label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12px; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
    )
)

# Add to map
site_map.add_child(distance_marker)
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
```

```python
# Launch Site Coordinates (Example: CCAFS SLC-40)
launch_site_lat, launch_site_lon = 28.563197, -80.576820  

# Closest Proximities (Replace with actual values)
city_lat, city_lon = 28.3922, -80.6077   # Example city coordinate
railway_lat, railway_lon = 28.5721, -80.5853  # Example railway coordinate
highway_lat, highway_lon = 28.5613, -80.5709  # Example highway coordinate
# Calculate distances
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)
# Function to add markers with distance labels
def add_marker_with_distance(map_obj, lat, lon, distance, label):
    marker = folium.Marker(
        [lat, lon],
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12px; color:#d35400;"><b>{}</b>: {:10.2f} KM</div>'.format(label, distance),
        )
    )
    map_obj.add_child(marker)

# Add markers for city, railway, and highway
add_marker_with_distance(site_map, city_lat, city_lon, distance_city, "City")
add_marker_with_distance(site_map, railway_lat, railway_lon, distance_railway, "Railway")
add_marker_with_distance(site_map, highway_lat, highway_lon, distance_highway, "Highway")
```

```python
# Draw lines for visualization
lines = folium.PolyLine(
    locations=[
        (launch_site_lat, launch_site_lon),
        (city_lat, city_lon),
        (railway_lat, railway_lon),
        (highway_lat, highway_lon)
    ],
    weight=2,
    color="blue",
    opacity=0.7
)

# Add to map
site_map.add_child(lines)
site_map
```
